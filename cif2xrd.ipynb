## pymatgenによるcifファイルの操作

- 用意するもの
    - pymatgen (uv pip install pymatgen)
    - matplotlib (uv add matplotlib)
    - cifファイル

- 何をするか？
    - 結晶構造分析にはcifファイル形式が用いられる
    - cifファイルから粉末結晶XRDデータを取得，描画するプログラムを作成する
    - ここでは，身近な物質を題材にpymatgenを動かしてみる

### 塩化ナトリウム(NaCl)の場合

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.io.cif import CifParser
import matplotlib.pyplot as plt
```

```python
# 塩化ナトリウム(NaCl)のcifファイルを取得
parse = "cif_data/1000041.cif" # 1000041: NaCl
parse = CifParser(parse)
parse = parse.parse_structures(primitive=False, symmetrized=False)[0] # 単位格子構造を保持

parse_xrd = XRDCalculator(wavelength='CuKa')

# XRDデータの描画
ax = parse_xrd.get_plot(parse)
ax.set_xlabel(r"2$\theta$[deg.]")
ax.set_ylabel("Intensity (scaled)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# XRDデータのピーク位置，強度，ミラー指数を取得
pattern = parse_xrd.get_pattern(parse)

print("<2-theta> <Intensity> <Miller indices (hkl)>")

for i in range(len(pattern.x[:3])): # 上位3個を表示
    two_theta = pattern.x[i]
    intensity = pattern.y[i]
    hkl = pattern.hkls[i][0]["hkl"]
    
    print(f"{two_theta:.4f} | {intensity:.4f} | {hkl}")
```
