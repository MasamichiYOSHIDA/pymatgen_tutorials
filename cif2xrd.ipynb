```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.io.cif import CifParser
```

```python
nacl = "cif_data/1000041.cif"
nacl = CifParser(nacl)
nacl = nacl.get_structures(primitive=False, symmetrized=False)[0]

nacl_xrd = XRDCalculator(wavelength='CuKa')
nacl_xrd.plot_structures(structures=[nacl])
```
